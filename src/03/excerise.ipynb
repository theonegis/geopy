```python
from pathlib import Path
from osgeo import gdal

gdal.UseExceptions()

root = Path('/Users/tanzhenyu/Dataware/GeoPy/MSI')
# 使用filter()过滤出root文件夹下的所有子文件夹
# 使用for循环遍历root文件夹下的所有子文件夹
for folder in filter(lambda p: p.is_dir(), root.iterdir()):
    # 每个子文件夹下是一景Landsat遥感影像，每个波段是一个GeoTIFF数据
    # 通过Path的glob()方法得到B1到B7波段文件，然后将波段数据叠加创建VRT图像
    inputs = [str(f) for f in folder.glob('*B[1-7].tif')]
    fn_vrt = f'{folder.stem}.vrt'
    gdal.BuildVRT(fn_vrt, inputs, 
                  options=gdal.BuildVRTOptions(separate=True))
    
    # 将多波段的VRT文件转为GeoTIFF格式
    vrt_ds = gdal.Open(fn_vrt)
    gdal.Translate(f'{folder.stem}.tif', vrt_ds)
    del vrt_ds
```
