```python
from pathlib import Path
from osgeo import ogr

ogr.UseExceptions()

fn = r"C:\Users\TheOne\Dataware\GeoPy\Shannxi\Shannxi.shp"
in_ds: ogr.DataSource = ogr.Open(fn)
in_lyr: ogr.Layer = in_ds.GetLayer()

output: str = 'Buffer.shp'
# 创建输出Buffer文件
driver: ogr.Driver = ogr.GetDriverByName('ESRI Shapefile')
if Path(output).exists():
    driver.DeleteDataSource(output)
# 新建DataSource，Layer
out_ds: ogr.DataSource = driver.CreateDataSource(output)
out_lyr: ogr.Layer = out_ds.CreateLayer(Path(output).stem,
                                        in_lyr.GetSpatialRef(), ogr.wkbPolygon)
def_feat: ogr.FeatureDefn = out_lyr.GetLayerDefn()

# 遍历原始的Shapefile文件给每个Geometry做Buffer操作
for feature in in_lyr:
    geometry = feature.GetGeometryRef()
    buffer = geometry.Buffer(6500)
    out_feat = ogr.Feature(def_feat)
    out_feat.SetGeometry(buffer)
    out_lyr.CreateFeature(out_feat)
    out_feat = None
out_ds.FlushCache()
del in_ds, out_ds
```

```python
from osgeo import ogr
ogr.UseExceptions()

fn1 = r'C:\Users\TheOne\Dataware\GeoPy\ShannXi\ShannXi.shp'
fn2 = r'C:\Users\TheOne\Dataware\GeoPy\Guanzhong\关中城市群.shp'

# 读取两个数据集
ds1, ds2 = ogr.Open(fn1), ogr.Open(fn2)
lyr1, lyr2 = ds1.GetLayer(), ds2.GetLayer()

output = []
# 通过两个for循环依次计算第一个数据集中的多边形和第二个数据集中多边形的交集部分，将Intersection的结果加入存储结果的列表中
# 如果我们要将结果存储为矢量数据文件，注意还需要在循环中读取第二个数据集中的多边形属性进行存储，然后将多边形及其属性同时写入矢量数据文件
for feat1 in lyr1:
    geom1 = feat1.GetGeometryRef()
    for feat2 in lyr2:
        geom2 = feat2.GetGeometryRef()
        if geom2.Intersects(geom1):
            output.append(geom2.Intersection(geom1))
    lyr2.ResetReading()

# 遍历存储结果的列表，输出多边形的WKT表示形式
for geom in output:
    print(geom.ExportToWkt())
```

```python
from osgeo import gdal, gnm

gdal.UseExceptions()
gnm.UseExceptions()

driver = gdal.GetDriverByName('GNMFile')
ds = driver.Create('.', 0, 0, 0, gdal.GDT_Unknown,
                   options=['net_name=well_pipe_network',
                            'net_description=GNM for wells and piples',
                            'net_srs=EPSG:4326'])

# 导入管道文件到网络模型文件
fn_pipes = '/Users/tanzhenyu/Dataware/GeoPy/gnm/pipes.shp'
ds_pipes = gdal.OpenEx(fn_pipes, gdal.OF_VECTOR)
lyr_pipes = ds_pipes.GetLayerByIndex(0)
ds.CopyLayer(lyr_pipes, 'pipes')
# 导入井文件到网络模型文件
fn_wells = '/Users/tanzhenyu/Dataware/GeoPy/gnm/wells.shp'
ds_wells = gdal.OpenEx(fn_wells, gdal.OF_VECTOR)
lyr_wells = ds_wells.GetLayerByIndex(0)
ds.CopyLayer(lyr_wells, 'wells')

dn = gnm.CastToGenericNetwork(ds)
print('构建网络拓扑关系...')
ret = dn.ConnectPointsByLines(['pipes', 'wells'], 1e-5, 1, 1, gnm.GNM_EDGE_DIR_BOTH)
assert ret == 0, '网络拓扑构建失败！'


del ds_pipes, ds_wells
del dn, ds
```

```python
from osgeo import gdal, ogr, gnm, osr

gdal.UseExceptions()
gnm.UseExceptions()

ds = gdal.OpenEx('./well_pipe_network')
dn = gnm.CastToGenericNetwork(ds)
print('计算两点之间的最短路径...')
result = dn.GetPath(40, 60, gnm.GATDijkstraShortestPath)
# 创建输出结果矢量文件
driver = ogr.GetDriverByName('GPKG')
output = '/Users/tanzhenyu/Dataware/GeoPy/gnm/path.gpkg'
dst = driver.CreateDataSource(output)
srs = osr.SpatialReference()
srs.ImportFromEPSG(4326)
lyr_p = dst.CreateLayer('Point', geom_type=ogr.wkbPoint, srs=srs)
lyr_l = dst.CreateLayer('Line', geom_type=ogr.wkbLineString, srs=srs)
lyr_p.CreateField(ogr.FieldDefn("ID", ogr.OFTInteger))
lyr_l.CreateField(ogr.FieldDefn("ID", ogr.OFTInteger))


for feat in result:
    geom = feat.GetGeometryRef()
    idx = feat.GetFieldAsString(0)
    print(geom.ExportToWkt())

    if geom.GetGeometryType() == ogr.wkbPoint:
        feature = ogr.Feature(lyr_p.GetLayerDefn())
        feature.SetGeometry(geom)
        feature.SetField("ID", idx)
        lyr_p.CreateFeature(feature)
    elif geom.GetGeometryType() == ogr.wkbLineString:
        feature = ogr.Feature(lyr_l.GetLayerDefn())
        feature.SetGeometry(geom)
        feature.SetField("ID", idx)
        lyr_l.CreateFeature(feature)

dst.FlushCache()
del ds, dn, dst
```

```python
from pathlib import Path
from osgeo import ogr

def create_ds(file_name, file_format, geom_type, srs, overwrite=False):
    driver = ogr.GetDriverByName(file_format)
    path  = Path(file_name)
    if path.exists() and overwrite is True:
        path.unlink()
    ds = driver.CreateDataSource(file_name)
    lyr = ds.CreateLayer(path.stem, srs, geom_type)
    return ds, lyr

def dissolve(input, output, multipoly=False, overwrite=False):
    ds = ogr.Open(input)
    lyr = ds.GetLayer()
    out_ds, out_lyr = create_ds(output, ds.GetDriver().GetName(), lyr.GetGeomType(), lyr.GetSpatialRef(), overwrite)
    defn = out_lyr.GetLayerDefn()
    multi = ogr.Geometry(ogr.wkbMultiPolygon)
    for feat in lyr:
        if feat.geometry():
            feat.geometry().CloseRings() # this copies the first point to the end
            wkt = feat.geometry().ExportToWkt()
            multi.AddGeometryDirectly(ogr.CreateGeometryFromWkt(wkt))
    union = multi.UnionCascaded()
    if multipoly is False:
        for geom in union:
            poly = ogr.CreateGeometryFromWkb(geom.ExportToWkb())
            feat = ogr.Feature(defn)
            feat.SetGeometry(poly)
            out_lyr.CreateFeature(feat)
    else:
        out_feat = ogr.Feature(defn)
        out_feat.SetGeometry(union)
        out_lyr.CreateFeature(out_feat)
        out_ds.Destroy()
    del ds
```
