```python
from pathlib import Path
from osgeo import ogr

ogr.UseExceptions()

# 读取被裁剪的数据
in_ds: ogr.DataSource = ogr.Open('/Users/tanzhenyu/Dataware/GeoPy/ShannXi/Rivers.shp')
in_lyr: ogr.Layer = in_ds.GetLayer()

# 读取裁剪范围数据
extent_ds: ogr.DataSource = ogr.Open('/Users/tanzhenyu/Dataware/GeoPy/XiAn/XiAn.shp')
extent_lyr: ogr.Layer = extent_ds.GetLayer()

out_fn: str = '/Users/tanzhenyu/Dataware/GeoPy/XiAn/Rivers.shp'
# 创建被裁剪以后的输出文件
driver: ogr.Driver = ogr.GetDriverByName('ESRI Shapefile')
if Path(out_fn).exists():
    driver.DeleteDataSource(out_fn)
# 新建DataSource，Layer
out_ds: ogr.DataSource = driver.CreateDataSource(out_fn)
out_lyr: ogr.Layer = out_ds.CreateLayer(out_fn.split('.')[0],
                                        in_lyr.GetSpatialRef(),
                                        in_lyr.GetGeomType())
# 开始进行裁剪（请确保投影一致）
in_lyr.Clip(extent_lyr, out_lyr)
out_ds.FlushCache()
del in_ds, extent_ds, out_ds
```

```python
from osgeo import ogr

ogr.UseExceptions()


# 读取输入数据
in_fn = r"C:\Users\TheOne\Dataware\GeoPy\ShannXi\ShaanxiCounty.shp"
in_ds: ogr.DataSource = ogr.Open(in_fn)
in_lyr: ogr.Layer = in_ds.GetLayer()

# 创建融合以后的输出文件
out_fn = 'Dissolved.shp'
driver: ogr.Driver = ogr.GetDriverByName('ESRI Shapefile')
# 新建输出DataSource，Layer
out_ds: ogr.DataSource = driver.CreateDataSource(out_fn)
out_lyr: ogr.Layer = out_ds.CreateLayer(out_fn.split('.')[0],
                                        in_lyr.GetSpatialRef(),
                                        in_lyr.GetGeomType())
defn: ogr.FeatureDefn = out_lyr.GetLayerDefn()
geoms: ogr.Geometry = ogr.Geometry(ogr.wkbMultiPolygon)

for feat in in_lyr:
    feat.geometry().CloseRings() # 强制多边形起点和终点闭合
    geoms.AddGeometry(feat.geometry())

# 使用UnionCascaded()将多个多边形进行融合，该方法要求输入数据集合体必须为MultiPolygon
union = geoms.UnionCascaded()
# 根据生成的Geometry几何体，新建Feature要素，并将其关联到输出图层
out_feat = ogr.Feature(defn)
out_feat.SetGeometry(union)
out_lyr.CreateFeature(out_feat)
out_ds.FlushCache()
del out_ds, in_ds
```
