```python
from osgeo import ogr
ogr.UseExceptions()

fn_province: str = r"C:\Users\TheOne\Dataware\GeoPy\China\Province.shp"
ds_province: ogr.DataSource = ogr.Open(fn_province)
lyr_province: ogr.Layer = ds_province.GetLayer()
# 使用filter()方法找出湖北省
ft_hubei = next(filter(lambda f: '湖北' in f.GetField('NAME'), lyr_province))

fn_city = r"C:\Users\TheOne\Dataware\GeoPy\China\Cities.shp"
ds_city = ogr.Open(fn_city)
lyr_city = ds_city.GetLayer()
# 使用filter()方法过滤出落在湖北省境内的所有市
selected = filter(lambda f: f.GetGeometryRef().Within(ft_hubei.GetGeometryRef()), lyr_city)
for city in selected:
    print(city.GetField('name'))
del ds_province
del ds_city
```

```python
from osgeo import ogr
ogr.UseExceptions()


fn: str = r"C:\Users\TheOne\Dataware\GeoPy\China\Cities.shp"
ds: ogr.DataSource = ogr.Open(fn)
cities: ogr.Layer = ds.GetLayer()
# 使用filter()方法找出武汉市
city: ogr.Feature = next(filter(lambda f: '武汉' in f.GetField('name'), cities))
# 调用ResetReading()方法特别重要，如果不ResetReading的话后面的对Feature的遍历会出错
cities.ResetReading()
# 根据每个市到武汉市的距离进行排序
selected = sorted(cities, key=lambda f: f.GetGeometryRef().Distance(city.GetGeometryRef()))
for i in range(1, 4):
    print(selected[i].GetField('name'))
del ds
```
