```python
def fibonacci(n):
    a, b, count = 0, 1, 0
    while True:
        if (count > n):
            return
        else:
            yield a
        a, b = b, a + b
        count += 1

f = fibonacci(5)

print(type(f))

for _ in range(5):
     print(next(f), end=' ')
```

```python
# 对上一节中使用next()函数遍历容器元素时可能出现的StopIteration异常进行处理
greeting = 'How are you'

it = iter(greeting)

while True:
    try:
        print(next(it), end='\t')
    except StopIteration:
        break

# 被除数为0异常，使用finally输出每次的运行结果
x = [i for i in range(-2, 2)]

y = []

for i in x:
         try:
             y.append(2 / i)
         except ZeroDivisionError as e:
             print(e)
         finally:
             print(y)
```

```python
# 若干个数字相加，我们可以在形参前面添加一个*号，表示不具名的多个参数

# 实际上传入函数的不定长参数被转为一个tuple，我们可以用for-in循环对传入的参数进行遍历

def add(*nums):
    # 这里nums形参是一个tuple
    result = 0
    for i in nums:
        result += i
    return result

print(add(1, 2, 3))

print(add(9, 1, 2, 8, 3, 6, 4))

# 如果是具名不定长参数，则可以在形参前面添加两个*号，实际上传入函数的所有参数被转为一个 dict
def info(**args):
    # 这里args形参是一个dict
    for k, v in args.items():
        print(f'{k} -> {v}')

info(name='小明', age=12, gender='M')
```

```python
from pathlib import Path

import shutil

d = Path('Music')

print(d.exists())

# 新建一个文件夹
d.mkdir()

p = Path('七里香.mp3')
print(p.exists())

# 使用shutil.copy(src, dst)函数进行文件拷贝，返回拷贝目标文件(字符串形式

# 这里将文件p拷贝到文件夹d
dest_file = shutil.copy(str(p), str(d))
print(dest_file)

# 文件another为拷贝到d文件夹下的p文件，“_”表示上一行代码的返回结果
another = Path(dest_file)

# 使用shutil.move(str(src), str(dst))函数进行文件移动或重命名，返回移动或重命名之后的目标文件。
# 若src所在目录和dst目录或dst文件所在目录为同一目录，则move()函数相当于进行重命名操作，否则为文件移动

# 这里将文件another重命名为“周杰伦-七里香.mp3”
new_name = d / '周杰伦-七里香.mp3'
shutil.move(str(another), str(new_name))

# 此时的another文件的名称为“周杰伦-七里香.mp3”
another = d / '周杰伦-七里香.mp3'
print(another)

# 使用Path类的unlink()方法可以进行文件删除
another.unlink()

print(another.exists())

# 使用shutil.copytree(str(src), str(dst))函数进行文件夹移动或重命名，使用和shutil.copy()方法类似

# 重命名文件夹d为“拷贝”
print(shutil.copytree(d, '拷贝'))

d2 = Path('拷贝')

print(d2.exists())

print(d2.is_dir())

# shutil.move()函数既可以用来移动或重命名文件也可以用来移动或重命名文件夹
dest_file2 = shutil.move(str(d2), str(d))
print(dest_file2)

d3 = Path(dest_file2)

# 移动之后到d2文件夹就不存在了
print(d2.exists())

print(d3.exists())

# 使用shutil.rmtree()函数删除文件夹
shutil.rmtree(str(d3))

print(d3.exists())
```

```python
import math

class Rectangle(object):
    """
    Python中使用class关键字定义类，后面是类名。类名后面的括号中是继承的父类
    如果没有合适的继承类，就使用object类，这是所有类的父类(关于继承我们会在后面进行详谈) 
    """
    
    def __init__(self, coordinates):
        """
        Python中以__开头和结尾的函数或变量是系统预定义的，具有特殊含义
        __init__()方法表示初始化该类的实例时需要执行的方法，第一个参数永远是self，表示创建的实例本身
        在__init__()方法内部，我们一般会把对象的属性绑定到self，已经进行一些其他初始化工作
        """
        width = math.sqrt(math.fabs(coordinates[0][0] - coordinates[1][0]) ** 2 +
                          math.fabs(coordinates[0][1] - coordinates[1][1]) ** 2)
        height = math.sqrt(math.fabs(coordinates[1][0] - coordinates[2][0]) ** 2 +
                           math.fabs(coordinates[1][1] - coordinates[2][1]) ** 2)

        self.coordinates = coordinates

        # 如果给定坐标不闭合，将多边形坐标闭合
        if self.coordinates[0] != self.coordinates[-1]:
            self.coordinates.append(self.coordinates[0])
    
        # 较长的边作为矩形的长，较短的作为矩形的宽 
        self.width = min(width, height) 
        self.height = max(width, height)

    def area(self):
        # 计算矩形面积
        return self.height * self.width
         
    def perimeter(self): # 计算矩形周长
        return 2 * (self.height + self.width)


# 定义了People类以后我们可以对类进行实例化生成对象
rect = Rectangle([(-1, 0), (0, 1), (1, 0), (0,-1)])
# 可以用isinstance(obj, cls)函数判断对象obj的原型是不是cls类 
print(isinstance(rect, Rectangle))

# 调用对象的属性和方法 
print(f'矩形长为{rect.height}，宽为{rect.width}')

print(f'矩形周长为{rect.perimeter()}，面积为{rect.area()}')
```

```python
import math

class Shape(object):
    def __init__(self, name, coordinates, fill='blank', stroke='black'):
        # 构造函数，对类进行初始化
        self.name = name # 多边形名称
        self.fill = fill # 多边形填充色
        self.stroke = stroke # 多边形边的颜色 
        self.coordinates = coordinates # 多边形坐标
        
    def area(self):
        pass
        
    def perimeter(self):
        pass


class Rectangle(Shape):
    # Rectangle继承自Shape
    def __init__(self, coordinates):
        # 使用super方法调用父类的构造函数
        super(Rectangle, self).__init__('Rectangle', coordinates) 
        # 使用assert断言判断坐标串个数为4，或者前后坐标串相等时为5
        assert (len(coordinates) == 4 or
                (len(coordinates) == 5 and self.coordinates[0] == self.coordinates[-1]))
        
        width = math.sqrt((coordinates[0][0] - coordinates[1][0]) ** 2 +
                          (coordinates[0][1] - coordinates[1][1]) ** 2)
        height = math.sqrt((coordinates[1][0] - coordinates[2][0]) ** 2 +
                           (coordinates[1][1] - coordinates[2][1]) ** 2)
        
        self.coordinates = coordinates
        # 如果给定坐标不闭合，将多边形坐标闭合
        if self.coordinates[0] != self.coordinates[-1]:
            self.coordinates.append(self.coordinates[0])
            
        # 较长的边作为多边形的长，较短的作为多边形的宽 
        self.width = min(width, height)
        self.height = max(width, height)
        
    def area(self):
        # 计算矩形面积
        return self.height * self.width
        
    def perimeter(self):
        # 计算矩形周长
        return 2 * (self.height + self.width)


class Circle(Shape):
    def __init__(self, coordinates, center=(0, 0), radius=0):
        super(Circle, self).__init__('Circle', coordinates)
        # 如果没有给定坐标串，给定圆心和半径也可以，因为圆心和半径可以唯一确定一个圆 
        if coordinates is None:
            self.center = center
            self.radius = radius
        else:
            # 圆心和圆上任意一点坐标也可以确定一个圆，所以给定坐标串个数为2 
            assert len(coordinates) == 2
            # coordinates为圆心坐标和圆上任意一点坐标
            self.center = coordinates[0] # 圆心坐标
            self.radius = math.sqrt((coordinates[0][0] - coordinates[1][0]) ** 2 +
                                       (coordinates[0][1] - coordinates[1][1]) ** 2)     
            
    def area(self):
        # 计算圆形面积
        return math.pi * self.radius ** 2

    def perimeter(self): 
        # 计算圆形周长
        return math.pi * self.radius * 2
```

```python
if __name__ == '__main__':
    coords = input('请输入几何图形坐标(如:1.2 2.5, 3.0 4.7, ...):')
    # 首先将coordinates以逗号分割开 coords.split(',')
    # 然后去除掉分割结果中前后的空格再以空格将坐标分割开p.strip().split(' ')
    # 最后将字符串坐标转为数字tuple([float(i) for i in p.strip().split(' ')])
    coords = [tuple([float(i) for i in p.strip().split(' ')]) for p in coords.split(',')]
    shape = None
    if len(coords) == 2:
        print(f'您输入圆形坐标为:{coords}')
        shape = Circle(coords)
    elif len(coords) == 4:
        print(f'您输入矩形坐标为:{coords}') 
        shape = Rectangle(coords)

    print(f'图形面积为:{shape.area()}') 
    print(f'图形周长为:{shape.perimeter()}')
```

```python
from functools import reduce
nums = filter(lambda x: isinstance(x, int), [1, 0, 1.0, 3.14, 'good'])

print(nums)

print([i for i in nums])

# 我们来看一个综合的案例，嵌套使用这三个函数实现复杂运算逻辑
print(reduce(lambda x, y: x + y, map(lambda x: (x + 1) ** 2,
                                filter(lambda x: isinstance(x, int),[1, 0, 1.0, 3.14, 'good']))))

res = 0

for i in [1, 0, 1.0, 3.14, 'good']:
    if isinstance(i, int):
        res += (i + 1) ** 2

print(res)
```
