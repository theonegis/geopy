```python
# struct模块按可以按照指定格式将字节流转换为Python指定的数据类型
# 本案例中我们通过借助struct木块的unpack()函数对字节流进行解析
import struct

file_name = '/Users/tanzhenyu/Dataware/400万/省会城市.shp'
with open(file_name, "rb") as file:
    # 读取前十个字节
    main_header = file.read(100)
    # 前4个字节是以大端存储的File Code，>表示的是大端存储，i表示的是无符号整型
    # 由于unpack()函数的返回值是一个元组，我们这里解析一个数据所以取出第一个元素即可
    file_code = struct.unpack(">i", main_header[:4])[0]
    # 第32到35个字节存储的是几何体的类型，<表示的是大端存储，i表示的是无符号整型
    shape_type = struct.unpack("<i", main_header[32:36])[0]
    print("File Code:", file_code)
    print("Shape Type:", shape_type)

    # 第36到67表示的是该文件存储数据的最大外接矩形
    x_min = struct.unpack("<d", main_header[36:44])[0]
    y_min = struct.unpack("<d", main_header[44:52])[0]
    x_max = struct.unpack("<d", main_header[52:60])[0]
    y_max = struct.unpack("<d", main_header[60:68])[0]
    print(f"Bounding Box: {(x_min, y_min, x_max, y_max)}")

    # 主头文件之后的字节表示的是几何体信息，每一条记录包含一个头文件和几何体坐标
    # 这里我们读取的是点文件，点的逻辑结构只有一个X和Y坐标
    while True:
            # 读取记录的头文件（8个字节）
            content_header = file.read(8)
            # 如果记录头文件为空，则说明文件已经读到了结尾
            if not content_header:
                break
            # 记录头文件前面使用两个无符号整型分别表示记录编号和内容长度
            record_number, content_length = struct.unpack(">2i", content_header)
            # 内容长度是按照16比特bit计算的，所以要乘以2转为字节byte长度
            print(f"Record {record_number}, Content Length: {content_length * 2} bytes")
            shape_data = file.read(content_length * 2)
            # 前四个字节存储的是几何体类型，后面的字节存储的才是点坐标，d表示的是浮点型
            x, y = struct.unpack("<2d", shape_data[4:21])
            print(f"Point: ({x}, {y})")
```
