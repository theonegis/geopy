```python
import numpy as np
import pandas as pd

# 由嵌套的list（二维数组）构造一个DataFrame
data = [[1, 2, 3],
        [4, 5, 6],
        [7, 8, 9]]
df = pd.DataFrame(data, index=[1, 2, 3],
                  columns=['a', 'b', 'c'])
print(df)
# 以MarkDown格式进行输出
print(df.to_markdown())

# 由dict构造一个DataFrame
data = {
    'a': [4, 5, 6],
    'b': [7, 8, 9],
    'c': [10, 11, 12]
}
df = pd.DataFrame(data, index=[1, 2, 3])
print(df)
```

```python
# 使用Pandas读取CSV数据
#  iris数据集是由三种鸢尾花，各50组数据构成的数据集
#  每个样本包含4个特征，分别为萼片(sepals)的长和宽、花瓣(petals)的长和宽
df = pd.read_csv('iris.csv')
# 查看数据前5行
print(df.head(5))
# 查看数据后5行
print(df.tail(5))
# 输出数据的行数和列数
print(df.shape)
```

```python
import numpy as np
import pandas as pd
# 按条件选择某些行(萼片宽度大于3.5的数据子集)
data = df[df.sepalwidth > 3.5]
# 输出所选数据的行数
print(len(data))
data = df[(df.sepalwidth > 4) & (df.petalwidth < 0.5)]
print(data)

# 选择数据表中的某些列
data = df[['petallength', 'petallength']]
print(data.head())

# 选择数据中的一列
data = df['petallength']
# 只有一列的数据表类型为Series
print(type(data))
# df['petallength']与df.petallength写法等价
print(np.alltrue(data == df.petallength))
```

```python
# 使用loc和iloc属性选择行或者列
# iloc使用基于整数索引Index的方式进行行或者列的选择
data = df.iloc[10: 16]
print(data)
data = df.iloc[:, [1, 2]]
print(data.head())

# loc使用基于列名或者基于列名的条件进行选择
data = df.loc[:, ['sepalwidth', 'petalwidth']]
print(data.head())
data = df.loc[df.sepalwidth > 3.5, ['sepalwidth', 'class']]
print(data.head())
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'Age': [24, 22, 23],
    'Gender': ['m', 'f', 'm'],
    'Name': ['张三', '李四', '王五']
})
print(df1)

df2 = pd.DataFrame({
    'Age': [20, 22, 21],
    'Gender': ['f', 'f', 'm'],
    'Name': ['Mary', 'Lisa', 'John']
})
print(df2)

# 将两个或多个DataFrame按行进行拼接
df = pd.concat([df1, df2])

print(df)
```

```python
df2 = pd.DataFrame({
    'From': ["Xi'an", 'Wuhan', 'Beijing'],
    'Major': ['GIS', 'RS', 'CS']
})
print(df2)

# 将两个或多个DataFrame按列进行拼接
df = pd.concat([df1, df2], axis=1)
print(df)
```

```python
# 给DataFrame增加一行
df.loc[len(df.index)] = [43, 'M', 'Jay', 'New Taipei', 'Music']
print(df)

import random
# 这里增加一列ID，这里的ID的取值为一个随机数和年龄的成绩（仅仅为了演示）
df['ID'] = df.Age * random.randint(0, 100)
print(df)
```

```python
# df1和df2具有相同的列名
df1 = pd.DataFrame({
    'x1': ['A', 'B', 'C'],
    'x2': [1, 2, 3]
})
print(df1.to_markdown())

df2 = pd.DataFrame({
    'x1': ['A', 'B', 'C', 'D'],
    'x2': [5, 2, 3, 4]
})
print(df2.to_markdown())

# 默认为内连接，使用内连接的结果类似求交集
df = pd.merge(df1, df2)
print(df.to_markdown())
# 使用外连接的结果类似与求并集
df = pd.merge(df1, df2, how='outer')
print(df.to_markdown())
```

```python
# 对于df1和df2列名不完全一样的情况
df1 = pd.DataFrame({
    'x1': ['A', 'B', 'C'],
    'x2': [1, 2, 3]
})
print(df1.to_markdown())

df2 = pd.DataFrame({
    'x1': ['A', 'B', 'D'],
    'x3': ['T', 'F', 'T']
})
print(df2.to_markdown())

# 分别使用左连接，右连接，内连接和外连接进行DataFrame的融合
left = pd.merge(df1, df2, how='left', on='x1')
print(left.to_markdown())
right = pd.merge(df1, df2, how='right', on='x1')
print(right.to_markdown())
inner = pd.merge(df1, df2, how='inner', on='x1')
print(inner.to_markdown())
outer = pd.merge(df1, df2, how='outer', on='x1')
print(outer.to_markdown())
```

```python
# 数据重复和缺失处理
import numpy as np
import pandas as pd

data = {
    "A": ["A", "B", "B", "C", "A"],
    "B": [50, 40, np.nan, 30, 50],
    "C": [True, False, False, False, True]
}

df = pd.DataFrame(data)
# 去掉重复的行
print(df.drop_duplicates())
# 去掉包含无效值的行
print(df.dropna())
# 将无效值进行填充
print(df.fillna(0))
```

```python
# 表格的行列转换（宽表格变长表格，长表格变宽表格）
import pandas as pd

data = {"Name": ["Pankaj", "Lisa", "David"],
        "ID": [1, 2, 3],
        "Role": ["CEO", "Editor", "Author"]}

df = pd.DataFrame(data)
print(df)
df_longer = pd.melt(df, id_vars=["ID"], value_vars=["Name", "Role"])
print(df_longer)

df_wilder = df_longer.pivot(index='ID', columns='variable', values='value')
print(df_wilder)
df_wilder = df_wilder.reset_index().rename_axis(None, axis=1)
print(df_wilder)
```
