```python
import numpy as np

# 这是一个三维数组，如果把多维数组看成一维数组的嵌套，第一个维度表示最外层的元素，以此类推，最后一个维度表示多维数组最内层的元素
nums = np.arange(18).reshape(3, 3, 2)
print(nums)

#每个维度的索引或切片用逗号分隔，切片通过start:stop:step表示
# stop默认值为-1，step默认值为1，默认值可以省略

# 访问第一个维度的第二个及其之后的元素，第二个维度的所有元素，第三位维度的第一个元素
print(nums[1:, :, 0])

# 访问第一个维度的第一个元素，第二个维度的第一和第二元素，第三个维度的最后一个元素
print(nums[0, :2, -1])

# 访问第一个维度的第一个和第二个元素，第三和第四维度的所有元素
# 对于后面的所有维度可以用省略号(三个点)表示
print(nums[:2, ...])

# 上面的写法等同于如下的写法
print(nums[:2, :, :])
```

```python
import numpy as np

# 生成一个2X2的数组，元素最小值为0，最大值为10
print('-' * 50)
x = np.random.randint(0, 10, (2, 2))
y = np.random.randint(0, 10, (2, 2))

# 使用pprint美化输出
from pprint import pprint

pprint(x)
pprint(y)

# 给每个元素进行加减乘除一个常数
pprint(x + 2)
pprint(x - 2.0)
pprint(x * 0.5)
pprint(x / 0.5)

# 数组的加减乘除(Element-wise，对应运算符作用在两个数组的对应元素上)
pprint(x + y)
pprint(x - y)
pprint(x * y)

# 对于除法，为了防止除数为0，我们给y添加一个微小的常数
pprint(x / (y + 1e-6))
```

```python
import numpy as np
from pprint import pprint

# 随机生成5X5的整数数组
nums = np.random.randint(0, 10, (5, 5))
pprint(nums)

# 使用np.where()方法得到满足条件的索引，where()函数的参数为判别条件，这里选择大于5的元素
indices = np.where(nums > 5)
pprint(indices)

# 输出对应元素值
print(nums[indices])

# 上述通过np.where()函数获得满足条件的索引然后输出对应元素这两个步骤可以通过np.extract()函数来实现，
# extract()函数第一个参数为判别条件，第二参数为多维数组
pprint(np.extract(nums > 5, nums))

# 选择满足判别条件的元素并重新赋值
# 给nums中大于5的元素赋值为5
nums[nums > 5] = 5
pprint(nums)
```

```python
import numpy as np
from pprint import pprint

nums = np.random.randint(-2, 10, 10)
pprint(nums)
# np.sort()函数返回输入数组的排序副本
# np.sort(a, axis=-1, kind=None, order=None)有四个参数，a为待排序的多维数组，axis指定排序的维度，
# kind指定排序算法，order在数组包含多个字段的情况下，指定排序算法使用的字段
pprint(np.sort(nums))

# 当数组为多维的时候，我们可以指定排序的维度
nums = np.random.randint(-2, 10, (4, 4))
pprint(nums)

# 第一个维度表示最外层的元素，以此类推，最后一个维度表示多维数组最内层的元素
pprint(np.sort(nums, 0))
pprint(np.sort(nums, 1))

# np.argsort()函数用于返回排好序以后的多维数组的元素在原始数组中的索引位置
# np.argsort()函数具有和np.sort()函数相同的参数
pprint(np.argsort(nums))

# 多维数组中指定维度的最大值，最小值
# 和np.amax(nums, 0)函数的输出一致
# 第一个维度(列)的最大值
pprint(nums.max(0))

# 和np.amin(nums, 1)函数的输出一致
# 第二个维度(行)的最小值
pprint(nums.min(1))

# 相应的，利用np.argmax()和np.argmin() 方法可以得到最大值和最小值对应的索引，大家可以观察一下
# 这个两个函数输出的索引是不是和max()和min()函数输出的最值相对应
pprint(np.argmax(nums, 0))
pprint(np.argmin(nums, 1))
```

```python
import numpy as np

# 将自定义函数转化为支持向量化编程的函数
def func(a, b):
    if a > b:
        return a - b
    else:
        return a + b

# 使用np.vectorize()函数将普通函数转化为支持向量化编程的函数
vfunc = np.vectorize(func)

# 随机生成两个二维数组
x = np.random.randint(0, 9, (3, 3))
print(x)

y = np.random.randint(0, 9, (3, 3))
print(y)

# 使用我们自定义的支持向量化的函数
# 注意fun(a, b)函数中a和b的维度要匹配
z = vfunc(x, y)
print(z)

# 参数a或者b如果是一个标量，维度也是匹配的，可以完成运算的
print(vfunc(x, 3))
```

```python
import pandas as pd

# 对于df1和df2列名不完全一样的情况 
df1 = pd.DataFrame({
       'x1': ['A', 'B', 'C'],
       'x2': [1, 2, 3]
})
print(df1.to_markdown())

df2 = pd.DataFrame({
       'x1': ['A', 'B', 'D'],
       'x3': ['T', 'F', 'T']
})
print(df2.to_markdown())

# 分别使用左连接，右连接，内连接和外连接进行DataFrame的融合
left = pd.merge(df1, df2, how='left', on='x1')

# 左连接保留了所有左边DataFrame的元素(nan表示空值)
print(left.to_markdown())

#右连接保留了所有右边DataFrame的元素
right = pd.merge(df1, df2, how='right', on='x1') 
print(right.to_markdown())

# 内连接保留所有公共元素
inner = pd.merge(df1, df2, how='inner', on='x1') 
print(inner.to_markdown())

# 外连接保留所有元素:
outer = pd.merge(df1, df2, how='outer', on='x1')
print(outer.to_markdown())
```

```python
from scipy import linalg
import numpy as np

# 方程系数
a = np.array([[7, 2], [4, 5]])
b = np.array([8, 10])

# 使用solve()函数求解线性方程组
res = linalg.solve(a, b)
print(res)

# 我们也可以通过求矩阵的逆来进行方程求解
# 注意这里要将np.ndarray转为np.mat进行矩阵运算
# np.mat(b).H为矩阵b的转置
res = np.mat(linalg.inv(a)) * np.mat(b).H
print(res)
```

```python
# 构建Delaunay三角网
import numpy as np
from scipy.spatial import Delaunay 
import matplotlib.pyplot as plt

# 随机生成20个2维点
points = np.random.rand(20, 2)
tri = Delaunay(points)

# 绘制三角格网
plt.triplot(points[:, 0], points[:, 1], tri.simplices)

# 绘制10这十个离散点
plt.plot(points[:,0], points[:,1], 'o')

plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)

plt.show()
```

```python
# 最简单的一个线性回归案例
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import (make_pipeline)

# 准备训练样本数据，构造一个x和y的关系，并给y添加噪声
x = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
param = np.array([1, 2])
y = np.dot(x, param) + np.random.rand(4)

# 对输入数据进行标准化，然后建立模型并进行训练
model = make_pipeline(StandardScaler(), LinearRegression())
model.fit(x, y)

# 进行模型精度评价(R2)
print(f'模型的决定系数为:{model.score(x, y)}:.2f')

# 进行预测
ins = np.array([[3, 5]])
pred = model.predict(ins)
print(f'输入数据{ins}, 得到预测结果为{pred}')
print(f'真实结果为{np.dot(ins, param)}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import measure

# 使用np.ogrid()函数构造一个二维数组
# 复数步长的设置是通过j进行设置的，如100j。复数前表示的是，用几个数值来等分整个区间。
x, y = np.ogrid[-np.pi: np.pi: 100j, -np.pi: np.pi: 100j]
r = np.sin(np.exp((np.sin(x) ** 3 + np.cos(y) ** 2)))

# 通过find_contours()函数寻找等高线，level参数指示了在图像中查找等高线的级别值
# 返回值为一个list，每个list中是等高线的坐标(row, column)
contours = measure.find_contours(r, level=0.8)

# 先以灰度图像显示图像r，然后使用for循环绘制寻找到的等高线
fig, ax = plt.subplots()
ax.imshow(r, cmap=plt.cm.gray)

for contour in contours:
    ax.plot(contour[:, 1], contour[:, 0], linewidth=2)

# 隐藏x和y坐标
ax.set_xticks([])
ax.set_yticks([])
plt.show()
```
